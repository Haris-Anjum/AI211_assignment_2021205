```python
# Python3 Program to print BFS traversal
# from a given source vertex. BFS(int s)
# traverses vertices reachable from s.
from collections import defaultdict

# This class represents a directed graph
# using adjacency list representation


class Graph:

	# Constructor
	def __init__(self):

		# default dictionary to store graph
		self.graph = defaultdict(list)

	# function to add an edge to graph
	def addEdge(self, u, v):
		self.graph[u].append(v)

	# Function to print a BFS of graph
	def BFS(self, s):

		# Mark all the vertices as not visited
		visited = [False] * (max(self.graph) + 1)

		# Create a queue for BFS
		queue = []

		# Mark the source node as
		# visited and enqueue it
		queue.append(s)
		visited[s] = True

		while queue:

			# Dequeue a vertex from
			# queue and print it
			s = queue.pop(0)
			print(s, end=" ")

			# Get all adjacent vertices of the
			# dequeued vertex s. If a adjacent
			# has not been visited, then mark it
			# visited and enqueue it
			for i in self.graph[s]:
				if visited[i] == False:
					queue.append(i)
					visited[i] = True

# Driver code


# Create a graph given in
# the above diagram
g = Graph()
g.addEdge(4, 2)
g.addEdge(4, 5)
g.addEdge(2, 8)
g.addEdge(2, 9)
g.addEdge(5, 2)
g.addEdge(5, 1)
g.addEdge(8, 9)
g.addEdge(9, 1)
g.addEdge(9, 5)

print("Following is Breadth First Traversal"
	" (starting from vertex 2)")
g.BFS(9)

# This code is contributed by Neelam Yadav
```

```python
# Python3 program to print DFS traversal
# from a given graph
from collections import defaultdict

# This class represents a directed graph using
# adjacency list representation


class Graph:

	# Constructor
	def __init__(self):

		# default dictionary to store graph
		self.graph = defaultdict(list)

	# function to add an edge to graph
	def addEdge(self, u, v):
		self.graph[u].append(v)

	# A function used by DFS
	def DFSUtil(self, v, visited):

		# Mark the current node as visited
		# and print it
		visited.add(v)
		print(v, end=' ')

		# Recur for all the vertices
		# adjacent to this vertex
		for neighbour in self.graph[v]:
			if neighbour not in visited:
				self.DFSUtil(neighbour, visited)

	# The function to do DFS traversal. It uses
	# recursive DFSUtil()
	def DFS(self, v):

		# Create a set to store visited vertices
		visited = set()

		# Call the recursive helper function
		# to print DFS traversal
		self.DFSUtil(v, visited)

# Driver's code


# Create a graph given
# in the above diagram
if __name__ == "__main__":
	g = Graph()
g.addEdge(4, 2)
g.addEdge(4, 5)
g.addEdge(2, 8)
g.addEdge(2, 9)
g.addEdge(5, 2)
g.addEdge(5, 1)
g.addEdge(8, 9)
g.addEdge(9, 1)
g.addEdge(9, 5)

print("Following is DFS from (starting from vertex 2)")
# Function call
g.DFS(9)

# This code is contributed by Neelam Yadav
```

```python
import heapq

def ucs(graph, start, goal):
    frontier = [(0, start, None)]  # Priority queue with start node
    explored = set()               # Set of explored nodes
    parents = {}                   # Dictionary of parent nodes
    
    while frontier:
        cost, node, parent = heapq.heappop(frontier)
        if node == goal:
            path = [node]
            while parent is not None:
                path.append(parent)
                parent = parents[parent]
            path.reverse()
            return cost, path
        explored.add(node)
        parents[node] = parent
        for neighbor, neighbor_cost in graph[node].items():
            if neighbor not in explored:
                heapq.heappush(frontier, (cost + neighbor_cost, neighbor, node))
    
    return None, None



start = 'a'
goal = 'c'

cost, path = ucs(graph, start, goal)
if path is not None:
    print(f"Cost: {cost}, Path: {' -> '.join(path)}")
else:
    print("No path found.")
```

```python
class Node:
    def __init__(self):
        self.arr = [[0 for j in range(8)] for i in range(8)]
        self.row = -1
        self.column = -1
        self.next = None


def print_node(a):
    for i in range(8):
        for j in range(8):
            print(a.arr[i][j], end=" ")
        print("\n")
    print("\n")


def is_valid(a):
    # vertical
    for i in range(a.row):
        if a.arr[i][a.column] == 1:
            return False
    # diagonal1
    for i, j in zip(range(a.row-1, -1, -1), range(a.column-1, -1, -1)):
        if a.arr[i][j] == 1:
            return False
    # diagonal2
    for i, j in zip(range(a.row-1, -1, -1), range(a.column+1, 8)):
        if a.arr[i][j] == 1:
            return False
    return True


def queen_8(a):
    last = None
    t = 1
    while a.row < 8:
        for i in range(8):
            b = Node()
            for j in range(8):
                for k in range(8):
                    b.arr[j][k] = a.arr[j][k]
            b.row = a.row + 1
            b.column = i
            b.arr[b.row][i] = 1
            if t == 1 and is_valid(b):
                a.next = b
                last = b
                last.next = None
                t = 0
            elif is_valid(b):
                last.next = b
                last = b
                last.next = None
        if a.next is not None:
            a = a.next
        if a.row == 7:
            print_node(a)
            return


if __name__ == '__main__':
    a = Node()
    queen_8(a)
```

```python
class Node:
    def __init__(self):
        self.arr = [[0 for j in range(8)] for i in range(8)]
        self.row = -1
        self.column = -1
        self.next = None


def print_node(a):
    for i in range(8):
        for j in range(8):
            print(a.arr[i][j], end=" ")
        print("\n")
    print("\n")


def is_valid(a):
    # vertical
    for i in range(a.row):
        if a.arr[i][a.column] == 1:
            return False
    # diagonal1
    for i, j in zip(range(a.row-1, -1, -1), range(a.column-1, -1, -1)):
        if a.arr[i][j] == 1:
            return False
    # diagonal2
    for i, j in zip(range(a.row-1, -1, -1), range(a.column+1, 8)):
        if a.arr[i][j] == 1:
            return False
    return True


def queen_8_bfs():
    start_node = Node()
    queue = [start_node]

    while queue:
        curr_node = queue.pop(0)

        # If all 8 queens have been placed, print the solution and return
        if curr_node.row == 7:
            print_node(curr_node)
            return

        # Otherwise, generate all valid moves from the current node and add them to the queue
        for i in range(8):
            new_node = Node()
            for j in range(8):
                for k in range(8):
                    new_node.arr[j][k] = curr_node.arr[j][k]

            new_node.row = curr_node.row + 1
            new_node.column = i
            new_node.arr[new_node.row][i] = 1

            if is_valid(new_node):
                curr_node.next = new_node
                new_node.next = None
                queue.append(new_node)


if __name__ == '__main__':
    queen_8_bfs()
```
